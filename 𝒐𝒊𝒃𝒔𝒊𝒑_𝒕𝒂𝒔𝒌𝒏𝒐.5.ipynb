```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Ayush Navneet Singh's data loading
ayush_ns_data = pd.read_csv('Advertising.csv', index_col=0)

# Ayush Navneet Singh's feature selection
ayush_ns_X = ayush_ns_data[['TV', 'Radio', 'Newspaper']]
ayush_ns_y = ayush_ns_data['Sales']

# Ayush Navneet Singh's train-test split
ayush_ns_X_train, ayush_ns_X_test, ayush_ns_y_train, ayush_ns_y_test = train_test_split(ayush_ns_X, ayush_ns_y, test_size=0.2, random_state=42)

# Ayush Navneet Singh's model training
ayush_ns_model = LinearRegression()
ayush_ns_model.fit(ayush_ns_X_train, ayush_ns_y_train)

# Ayush Navneet Singh's predictions
ayush_ns_predictions = ayush_ns_model.predict(ayush_ns_X_test)

# Ayush Navneet Singh's model evaluation
ayush_ns_mse = mean_squared_error(ayush_ns_y_test, ayush_ns_predictions)
ayush_ns_rmse = np.sqrt(ayush_ns_mse)
ayush_ns_r2 = r2_score(ayush_ns_y_test, ayush_ns_predictions)

print("Ayush Navneet Singh's Sales Prediction Results:")
print(f"Root Mean Squared Error: {ayush_ns_rmse:.2f}")
print(f"R-squared Score: {ayush_ns_r2:.2f}")

# Ayush Navneet Singh's coefficient analysis
ayush_ns_coef = pd.DataFrame({
    'Feature': ayush_ns_X.columns,
    'Coefficient': ayush_ns_model.coef_
})
print("\nAyush Navneet Singh's Coefficient Analysis:")
print(ayush_ns_coef)

# Ayush Navneet Singh's sample prediction
ayush_ns_sample = ayush_ns_X_test.iloc[0]
ayush_ns_predicted_sales = ayush_ns_model.predict([ayush_ns_sample])[0]
ayush_ns_actual_sales = ayush_ns_y_test.iloc[0]

print("\nAyush Navneet Singh's Sample Prediction:")
print(f"Advertising Spend: TV={ayush_ns_sample['TV']:.2f}, Radio={ayush_ns_sample['Radio']:.2f}, Newspaper={ayush_ns_sample['Newspaper']:.2f}")
print(f"Predicted Sales: {ayush_ns_predicted_sales:.2f}")
print(f"Actual Sales: {ayush_ns_actual_sales:.2f}")
print(f"Difference: {abs(ayush_ns_predicted_sales - ayush_ns_actual_sales):.2f}")

# Ayush Navneet Singh's visualization
plt.figure(figsize=(10, 6))
plt.scatter(ayush_ns_y_test, ayush_ns_predictions)
plt.plot([ayush_ns_y_test.min(), ayush_ns_y_test.max()], [ayush_ns_y_test.min(), ayush_ns_y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Ayush Navneet Singh's Sales Prediction: Actual vs Predicted")
plt.show()

# Ayush Navneet Singh's feature importance visualization
plt.figure(figsize=(10, 6))
ayush_ns_coef.plot(x='Feature', y='Coefficient', kind='bar')
plt.title("Ayush Navneet Singh's Feature Importance in Sales Prediction")
plt.xlabel("Advertising Medium")
plt.ylabel("Coefficient (Impact on Sales)")
plt.show()
```
