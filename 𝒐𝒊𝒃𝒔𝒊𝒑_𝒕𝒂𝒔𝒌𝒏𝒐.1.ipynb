```python

# Step 1: Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Loading Ayush's Iris dataset
data_ayush = pd.read_csv('Iris.csv')

# Step 3: Data preprocessing
# Renaming columns for Ayush-style clarity
data_ayush.columns = ['Ayush_ID', 'Ayush_SepalLength', 'Ayush_SepalWidth', 'Ayush_PetalLength', 'Ayush_PetalWidth', 'Ayush_Species']

# Dropping the ID column
data_ayush.drop('Ayush_ID', axis=1, inplace=True)

# Step 4: Splitting the dataset
X_ayush = data_ayush[['Ayush_SepalLength', 'Ayush_SepalWidth', 'Ayush_PetalLength', 'Ayush_PetalWidth']]
y_ayush = data_ayush['Ayush_Species']

# 70% training and 30% testing
X_train_ayush, X_test_ayush, y_train_ayush, y_test_ayush = train_test_split(X_ayush, y_ayush, test_size=0.6, random_state=42)

# Step 5: Feature scaling
from sklearn.preprocessing import StandardScaler
scaler_ayush = StandardScaler()
X_train_ayush = scaler_ayush.fit_transform(X_train_ayush)
X_test_ayush = scaler_ayush.transform(X_test_ayush)

ayush_classifier = RandomForestClassifier(
    n_estimators=5, 
    max_depth=2,      
    min_samples_split=3,  
    random_state=42
)
ayush_classifier.fit(X_train_ayush, y_train_ayush)

# Step 7: Predictions
y_pred_ayush = ayush_classifier.predict(X_test_ayush)

# Step 8: Confusion Matrix and Accuracy
cm_ayush = confusion_matrix(y_test_ayush, y_pred_ayush)
accuracy_ayush = accuracy_score(y_test_ayush, y_pred_ayush)

# Step 9: Displaying results
print(f"Ayush's Model Accuracy: {accuracy_ayush * 100:.2f}%")

# Confusion matrix visualization
plt.figure(figsize=(8,6))
sns.heatmap(cm_ayush, annot=True, fmt='d', cmap='Blues', xticklabels=ayush_classifier.classes_, yticklabels=ayush_classifier.classes_)
plt.title("Ayush's Confusion Matrix")
plt.ylabel('Actual Species')
plt.xlabel('Predicted Species')
plt.show()
```
